# **IMPORTANDO BIBLIOTECAS**

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def extrair_dados():
    url = "https://books.toscrape.com/catalogue/page-1.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    livros = []

    for item in soup.find_all('article', class_='product_pod'):
        titulo = item.h3.a['title']
        preco = item.find('p', class_='price_color').text
        avaliacao = item.p['class'][1]

        livros.append([titulo, preco, avaliacao])

    return pd.DataFrame(livros, columns=['Título', 'Preco', 'Avaliação'])

print("dados extraidos")
```

Extraindo Dados com Funçoes

```python
df_livros = extrair_dados()
print (df_livros)
```

# **Conversão de Tipo**

A conversao de tipo é necessaria, pois objetos do tipo string, impedem operações aritimeticas

```python
#Limpeza do preço
df_livros['Preco'] = df_livros['Preco'].astype(str).str.replace('£', '').astype(float)
```

```python
df_livros
```

# **Comando .map**

```python
mapa_notas = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
df_livros['Nota'] = df_livros['Avaliação'].map(mapa_notas)
```

`dropna()`

```python
#selecionando o que importa
df_livros_ml = df_livros[['Preco', 'Nota']].dropna()
```

```python
df_livros_ml_reindexado = df_livros_ml.reset_index(drop=True)
display(df_livros_ml_reindexado)
```
