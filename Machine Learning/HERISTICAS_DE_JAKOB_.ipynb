## Apresentação das Heurísticas de Jakob Nielsen com códigos em Python

### 1: Introdução

**Título:** As 10 Heurísticas de Usabilidade de Jakob Nielsen

**Conteúdo:**

* Criadas por Jakob Nielsen e Rolf Molich nos anos 90.
* Princípios gerais para design de interfaces de usuário eficazes e amigáveis.
* Continuam relevantes para avaliar e melhorar a usabilidade de sistemas interativos.
* Vamos explorar cada heurística com exemplos e uma breve demonstração em Python.

###2: Visibilidade do Status do Sistema

**Heurística:** O sistema deve sempre manter os usuários informados sobre o que está acontecendo, através de *feedback* apropriado dentro de um tempo razoável.

**Exemplo Conceitual:** Uma barra de progresso ao carregar um arquivo ou realizar uma operação demorada.

**Exemplo em Python:**

```
import time

def processar_dados(tempo_total=5):
    for i in range(tempo_total + 1):
        porcentagem = (i / tempo_total) * 100
        print(f"Processando... {porcentagem:.0f}% completo")
        time.sleep(1)
    print("Processamento concluído!")

processar_dados()
```

**Explicação:** O código simula um processo e informa ao usuário o progresso em porcentagem a cada segundo.

### 3: Correspondência entre o Sistema e o Mundo Real

**Heurística:** O sistema deve falar a linguagem dos usuários, com palavras, frases e conceitos familiares a eles, em vez de termos internos do sistema. Deve seguir convenções do mundo real, tornando a informação natural e intuitiva.

**Exemplo Conceitual:** Usar ícones de "salvar" que se assemelham a disquetes (mesmo que o sistema não use disquetes internamente).

**Exemplo em Python:**

```
acoes = {"abrir": "Abrir um arquivo existente",
         "salvar": "Salvar o arquivo atual",
         "imprimir": "Imprimir o documento"}

def exibir_ajuda(acao):
    if acao in acoes:
        print(f"{acao.capitalize()}: {acoes[acao]}")
    else:
        print("Ação desconhecida.")

exibir_ajuda("salvar")
exibir_ajuda("exportar")
```

**Explicação:** O dicionário `acoes` usa termos comuns para descrever funcionalidades, facilitando a compreensão pelo usuário.

### 4: Controle e Liberdade do Usuário

**Heurística:** Os usuários frequentemente escolhem funções do sistema por engano e precisarão de uma "saída de emergência" claramente marcada para deixar o estado indesejado sem ter que passar por um diálogo estendido. Suporte para desfazer e refazer.

**Exemplo Conceitual:** Botões de "Desfazer" e "Refazer" em editores de texto.

**Exemplo em Python:**

```
historico = []

def adicionar_ao_historico(item):
    historico.append(item)
    print(f"Adicionado: {item}. Histórico: {historico}")

def desfazer():
    if historico:
        removido = historico.pop()
        print(f"Desfeito: {removido}. Histórico: {historico}")
    else:
        print("Nada para desfazer.")

adicionar_ao_historico("Texto A")
adicionar_ao_historico("Texto B")
desfazer()
adicionar_ao_historico("Texto C")
```

**Explicação:** A lista `historico` simula um histórico de ações, permitindo desfazer a última ação.

### 5: Consistência e Padrões

**Heurística:** Os usuários não deveriam ter que se perguntar se diferentes palavras, situações ou ações significam a mesma coisa. Siga convenções de plataforma e indústria.

**Exemplo Conceitual:** Usar consistentemente o mesmo ícone para a função de "excluir" em toda a interface.

**Exemplo em Python:**

```
def mostrar_mensagem_erro(tipo, mensagem):
    if tipo == "conexao":
        print("[ERRO] Falha na conexão: " + mensagem)
    elif tipo == "arquivo":
        print("[ERRO] Erro de arquivo: " + mensagem)
    else:
        print("[ERRO] " + mensagem)

mostrar_mensagem_erro("conexao", "Servidor não encontrado.")
mostrar_mensagem_erro("arquivo", "Arquivo inexistente.")
```

**Explicação:** A função `mostrar_mensagem_erro` usa um formato consistente para exibir mensagens de erro, facilitando a identificação do tipo de problema.

### 6: Prevenção de Erros

**Heurística:** Melhor do que boas mensagens de erro é um design cuidadoso que evita que os erros ocorram em primeiro lugar. Elimine condições propensas a erros ou verifique-as e ofereça aos usuários uma opção de confirmação antes de cometer a ação.

**Exemplo Conceitual:** Desabilitar botões que não estão disponíveis no contexto atual.

**Exemplo em Python:**

```
def realizar_acao(acao, estado):
    if acao == "salvar" and estado == "vazio":
        print("Aviso: Nenhum conteúdo para salvar.")
        return
    print(f"Ação '{acao}' realizada.")

realizar_acao("salvar", "vazio")
realizar_acao("salvar", "com_conteudo")
```

**Explicação:** A função `realizar_acao` verifica o estado do sistema antes de permitir a ação "salvar", prevenindo um erro em um estado vazio.

### 7: Reconhecimento em vez de Memorização

**Heurística:** Minimize a carga de memória do usuário tornando objetos, ações e opções visíveis. O usuário não deve ter que se lembrar de informações de uma parte do diálogo para outra. As instruções para o uso do sistema devem ser visíveis ou facilmente recuperáveis sempre que necessário.

**Exemplo Conceitual:** Exibir uma lista de arquivos recentes em vez de exigir que o usuário se lembre do nome do arquivo.

**Exemplo em Python:**

```
arquivos_recentes = ["documento1.txt", "imagem.png", "relatorio_final.pdf"]

def exibir_arquivos_recentes():
    print("Arquivos Recentes:")
    for i, arquivo in enumerate(arquivos_recentes):
        print(f"{i+1}. {arquivo}")

exibir_arquivos_recentes()
```

**Explicação:** A função `exibir_arquivos_recentes` mostra uma lista de arquivos recentes, eliminando a necessidade de o usuário se lembrar dos nomes.

### 8: Flexibilidade e Eficiência de Uso

**Heurística:** Aceleradores — invisíveis para o usuário novato — podem frequentemente acelerar a interação para o usuário especialista de tal forma que o sistema possa atender tanto a usuários novatos quanto experientes. Permita que os usuários personalizem ações frequentes.

**Exemplo Conceitual:** Atalhos de teclado (Ctrl+S para salvar, Ctrl+C para copiar).

**Exemplo em Python:**

```
atalhos = {"ctrl+s": "Salvar", "ctrl+c": "Copiar", "ctrl+v": "Colar"}

def interpretar_comando(comando):
    if comando in atalhos:
        print(f"Comando '{comando}' interpretado como: {atalhos[comando]}")
    else:
        print("Comando desconhecido.")

interpretar_comando("ctrl+s")
interpretar_comando("ctrl+x")
```

**Explicação:** O dicionário `atalhos` mapeia combinações de teclas a ações, oferecendo uma forma mais rápida de interação para usuários experientes.

### 9: Design Estético e Minimalista

**Heurística:** Cada informação extra em um diálogo compete com as unidades de informação relevantes e diminui sua visibilidade relativa.

**Exemplo Conceitual:** Uma interface limpa e organizada com apenas os elementos essenciais para a tarefa em mãos.

**Exemplo em Python (simulação de interface simplificada):**

```
def exibir_detalhes_produto(nome, preco):
    print(f"Produto: {nome}")
    print(f"Preço: R$ {preco:.2f}")

def exibir_detalhes_produto_completo(nome, preco, descricao, avaliacao):
    print(f"Produto: {nome}")
    print(f"Preço: R$ {preco:.2f}")
    print(f"Descrição: {descricao}")
    print(f"Avaliação: {avaliacao}/5")

print("--- Versão Minimalista ---")
exibir_detalhes_produto("Notebook", 2500.00)

print("\n--- Versão Detalhada ---")
exibir_detalhes_produto_completo("Notebook", 2500.00, "Ótimo para trabalho e estudo.", 4.5)
```

**Explicação:** A primeira função (`exibir_detalhes_produto`) apresenta apenas as informações essenciais, enquanto a segunda (`exibir_detalhes_produto_completo`) inclui detalhes adicionais. Em muitos casos, menos informação inicial é melhor.

### 10: Ajuda e Documentação

**Heurística:** Mesmo que o sistema possa ser usado sem documentação, pode ser necessário fornecer ajuda e documentação. Qualquer informação de ajuda deve ser fácil de pesquisar, focada na tarefa do usuário, listar etapas concretas a serem realizadas e não ser muito extensa.

**Exemplo Conceitual:** Uma seção de "Ajuda" bem organizada com perguntas frequentes e tutoriais.

**Exemplo em Python (simulação de um sistema de ajuda):**

```
ajuda = {
    "salvar": "Para salvar seu trabalho, clique no ícone de disquete ou pressione Ctrl+S.",
    "abrir": "Para abrir um arquivo existente, vá em 'Arquivo' > 'Abrir' ou clique no ícone de pasta."
}

def obter_ajuda(topico):
    if topico in ajuda:
        print(f"Ajuda para '{topico}': {ajuda[topico]}")
    else:
        print("Tópico de ajuda não encontrado.")

obter_ajuda("salvar")
obter_ajuda("exportar")
```

**Explicação:** O dicionário `ajuda` simula um sistema de ajuda onde o usuário pode obter informações específicas sobre funcionalidades.
