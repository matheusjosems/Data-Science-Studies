```python
import requests
```

```python
from  bs4 import BeautifulSoup
```

```python
pagina = requests.get('https://quotes.toscrape.com/')
```

```python
pagina
```

```python
print(pagina)
```

```python
dados_pagina = BeautifulSoup (pagina.text, 'html.parser')
#pegar o html e tranforma em algo que pode ser lido em python, ou seja, tranforma em objetos, no caso 'dados_pagina'
```

```python
print(dados_pagina.prettify())
```

```python
todas_frases= dados_pagina.find_all('div', class_='quote')
# class_ é diferente de class
```

```python
for div in todas_frases:
  print(div)
```

```python
for div in todas_frases:
  print(div.text)
```

```python
for div in todas_frases:
  texto = div.find('span', class_='text').text
  print(texto)
```

```python
import pandas as pd
```

```python
lista_frases = []

for div in todas_frases:
  quote_text = div.find('span', class_='text').text
  author = div.find('small', class_='author').text
  tags = [tag.text for tag in div.find('div', class_='tags').find_all('a', class_='tag')]
  lista_frases.append({
      'quote': quote_text,
      'author': author,
      'tag': tags
      })
frases = pd.DataFrame(lista_frases)
```
