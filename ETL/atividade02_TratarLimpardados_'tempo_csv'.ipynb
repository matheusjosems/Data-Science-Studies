```python
import pandas as pd
import seaborn as srn
import statistics  as sts
```

```python
dataset = pd.read_csv("tempo.csv", sep=";")
dataset.head()
```

```python
agrupadoAparencia = dataset.groupby(['Aparencia']).size()
agrupadoAparencia
```

```python
agrupadoTemperatura = dataset.groupby(['Temperatura']).size()
agrupadoTemperatura
```

```python
#estar entre 0 ~ 100
agrupamentoUmidade = dataset.groupby(['Umidade']).size()
agrupamentoUmidade
```

```python
#Vento
agrupamentoVento = dataset.groupby(['Vento']).size()
agrupamentoVento
```

```python
#Jogar
agrupamentoJogar = dataset.groupby(['Jogar']).size()
agrupamentoJogar
```

```python
dataset.isnull().sum()
```

```python
dataset['Umidade'] = dataset['Umidade'].astype(float)
agrupamentoUmidade
```

```python
dataset['Temperatura'] = dataset['Temperatura'].astype(float)
agrupadoTemperatura
```

```python
dataset.head()
```

```python
# Aparencia Temperatura	Umidade	Vento	Jogar

dataset['Umidade'].describe()
```
