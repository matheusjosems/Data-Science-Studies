```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv("dados.csv", sep=";")
dados.head()
```

```python
dados.shape
```

```python
sns.histplot(dados.PIB, kde=True, bins=10, color='blue').set(title='PIB')
```

```python
sns.histplot(dados.VALOREMPENHO, kde=True, bins=10, color='blue').set(title="VALOR EMPENHO")
```

```python
agrupado = dados.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
```

```python
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray')
```

```python
agrupado2 = dados.sort_values('VALOREMPENHO').head(10)
agrupado2 =agrupado2.iloc[:,[1,3]]
agrupado2
```

```python
agrupado2.plot.bar(x = 'MUNICIPIO', y = 'VALOREMPENHO', color= 'gray')
```
