## Q1 — Fácil | Contagem de Vogais

Leia o enunciado com atenção antes de começar.

Implemente a função `contar_vogais(texto)` que receba uma string e retorne um dicionário com a contagem de cada vogal (`a`, `e`, `i`, `o`, `u`) presente no texto. A função deve:
- Ignorar maiúsculas e minúsculas
- Tratar vogais acentuadas como sua vogal base (`'á'`, `'à'`, `'ã'` → contam como `'a'`; `'é'`, `'ê'` → `'e'`; e assim por diante)
- Retornar o dicionário com todas as cinco vogais, mesmo que a contagem seja zero

**Exemplo esperado:**
```
contar_vogais("Ouvidoria Geral da União")
→ {'a': 3, 'e': 1, 'i': 2, 'o': 2, 'u': 2}
```

````python
# SUA RESPOSTA AQUI — Q1

def contar_vogais(texto):

  vogais= {'a':0,'e':0,'i':0,'o':0,'u':0}

  for char in texto:
      char = char.lower()
      if char in 'aáàãâ':
          vogais['a'] += 1
      elif char in 'eéê':
          vogais['e'] += 1
      elif char in 'ií':
          vogais['i']  += 1
      elif char in 'oóôõ':
          vogais['o'] += 1
      elif char in 'uú':
            vogais['u'] += 1
  return vogais


# Testes
print(contar_vogais("Ouvidoria Geral da União"))
# Esperado: {'a': 3, 'e': 1, 'i': 2, 'o': 2, 'u': 2}

print(contar_vogais("CGU"))
# Esperado: {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 1}
````

## Q2 — Fácil | Número por Extenso

Leia o enunciado com atenção antes de começar.

Implemente a função `por_extenso(n)` que receba um número inteiro entre `0` e `99` e retorne esse número escrito por extenso em português.

**Exemplos esperados:**
```
por_extenso(0)  → 'zero'
por_extenso(7)  → 'sete'
por_extenso(15) → 'quinze'
por_extenso(21) → 'vinte e um'
por_extenso(42) → 'quarenta e dois'
por_extenso(99) → 'noventa e nove'
```

````python
# SUA RESPOSTA AQUI — Q2

def por_extenso(n):

    numeros = {0:'zero', 1:'um', 2:'dois', 3:'três', 4:'quatro',
    5:'cinco', 6:'seis', 7:'sete', 8:'oito', 9:'nove', 10:'dez',
    11:'onze', 12:'doze', 13:'treze',14:'quatorze', 15:'quinze', 16:'dezesseis',17:'dezessete', 18:'dezoito', 19:'dezenove',
    20:'vinte', 30:'trinta', 40:'quarenta',
    50:'cinquenta', 60:'sessenta', 70:'setenta',
    80:'oitenta', 90:'noventa'}


    if n<0 or n>99 :
      print("ERRO: esse número não está entre 0 e 99")
      return None

    if n in numeros:
      return numeros [n]
    else:
      dezenas = (n // 10 ) * 10
      unidade = n % 10

    return numeros[dezenas] + ' e ' + numeros[unidade]

# Testes
for n in [0, 7, 15, 21, 42, 99]:
    print(f"{n:2d} → {por_extenso(n)}")
````

## Q3 — Média | Anagrama e Frequência de Palavras

Leia o enunciado com atenção antes de começar. Esta questão tem duas partes independentes.

**Parte A — Anagrama:**
Implemente `sao_anagramas(s1, s2)` que retorne `True` se as duas strings forem anagramas entre si (contêm exatamente as mesmas letras com as mesmas frequências), ignorando espaços e maiúsculas.

**Parte B — Frequência de palavras:**
Implemente `frequencia_palavras(texto)` que retorne um dicionário com a contagem de cada palavra no texto, ordenado da palavra mais frequente para a menos frequente.

**Exemplos esperados:**
```
sao_anagramas('Roma', 'amor')     → True
sao_anagramas('listen', 'silent') → True
sao_anagramas('hello', 'world')   → False

frequencia_palavras("o gato viu o rato e o rato fugiu")
→ {'o': 3, 'rato': 2, 'gato': 1, 'viu': 1, 'e': 1, 'fugiu': 1}
```

````python
# SUA RESPOSTA AQUI — Q3

def sao_anagramas(s1, s2):
    s1= s1.lower().replace(" ","")
    s2= s2.lower().replace(" ","")
    palavras_iguais = True

    if len(s1) != len(s2):
      palavras_iguais = False

    letras_s1 = {}
    letras_s2 = {}

    for letra in s1:
      if letra in letras_s1:
        letras_s1[letra] += 1
      else:
        letras_s1[letra] =1

    for letra in s2:
      if letra in letras_s2:
        letras_s2[letra] += 1
      else:
        letras_s2[letra] =1

    if letras_s1 != letras_s2:
      palavras_iguais= False

    return palavras_iguais

def frequencia_palavras(texto):
    palavra_lista = texto.lower().split()
    frequencia ={}
    resultado=[]

    for palavra in palavra_lista:

      if palavra in frequencia:
        frequencia [palavra]+=1
      else:
        frequencia[palavra] = 1

    for palavra in frequencia :
      resultado.append([palavra, frequencia[palavra]])

    lista = resultado


    for i in range(len(lista)):
      for j in range(len(lista)-1):
        if lista[j][1]< lista [j+1][1]:
          troca = lista [j]
          lista[j] = lista [j+1]
          lista [j+1] = troca

    return dict(lista)


# Testes
print(sao_anagramas('Roma', 'amor'))       # True
print(sao_anagramas('listen', 'silent'))   # True
print(sao_anagramas('hello', 'world'))     # False
print(frequencia_palavras("o gato viu o rato e o rato fugiu"))
````

## Q4 — Média | Ordenação e Mediana

Leia o enunciado com atenção antes de começar.

**Parte A:** implemente o algoritmo **Insertion Sort** na função `insertion_sort(lista)`, que deve retornar uma nova lista com os elementos em ordem crescente. Não modifique a lista original.

**Parte B:** usando sua função `insertion_sort`, implemente `mediana(lista)` que calcule a mediana dos valores. Lembre-se: se o número de elementos for par, a mediana é a média dos dois valores centrais.

**Exemplos esperados:**
```
insertion_sort([5, 2, 8, 1, 9, 3]) → [1, 2, 3, 5, 8, 9]
mediana([5, 2, 8, 1, 9, 3])        → 4.0
mediana([7, 1, 4])                 → 4
```

````python
# SUA RESPOSTA AQUI — Q4

def insertion_sort(lista):

  lista_ordenada = lista.copy()

  for posicao_atual in range(1, len(lista_ordenada)):
    numero_escolhido = lista_ordenada[posicao_atual]
    posicao_anterior = posicao_atual - 1

    while posicao_anterior >= 0 and lista_ordenada[posicao_anterior]> numero_escolhido:
      lista_ordenada[posicao_anterior +1]= lista_ordenada[posicao_anterior]
      posicao_anterior-=1

    lista_ordenada[posicao_anterior+ 1 ] = numero_escolhido

  return lista_ordenada

def mediana(lista):

  lista_ordenada = insertion_sort(lista)

  total_numeros =len(lista_ordenada)

  if total_numeros % 2 != 0:
    posicao_meio = total_numeros // 2
    mediana = lista_ordenada[posicao_meio]

  else:
    meio1 = (total_numeros // 2) -1
    meio2 = total_numeros // 2

    mediana = (lista_ordenada[meio1]+lista_ordenada[meio2])/2

  return mediana

# Testes
print(insertion_sort([5, 2, 8, 1, 9, 3]))  # [1, 2, 3, 5, 8, 9]
print(mediana([5, 2, 8, 1, 9, 3]))          # 4.0
print(mediana([7, 1, 4]))                   # 4
````

## Q5 — Desafio | Menor Caminho em Grade

Leia o enunciado com atenção antes de começar. Esta é a questão mais difícil do bloco — se travar, siga em frente e volte depois.

Você recebe uma grade `N × M` de inteiros, onde cada célula representa o custo de passar por ela. Partindo do canto superior esquerdo `(0, 0)` e chegando ao canto inferior direito `(N-1, M-1)`, você só pode se mover para a **direita** ou para **baixo** a cada passo.

Implemente `menor_custo(grade)` que retorne o **menor custo total** do caminho, somando todas as células percorridas (incluindo início e fim).

**Exemplo:**
```
grade = [
    [1, 3, 1],
    [1, 5, 1],
    [4, 2, 1]
]
menor_custo(grade) → 7   (caminho: 1 → 3 → 1 → 1 → 1)
```

> Dica: pense em como o menor custo para chegar em `(i, j)` depende dos custos de `(i-1, j)` e `(i, j-1)`. Isso leva a uma solução eficiente sem testar todos os caminhos possíveis.

````python
# SUA RESPOSTA AQUI — Q5

def menor_custo(grade):

  quantidade_linhas = len(grade)
  quantidade_colunas = len(grade[0])
  custos_acumulados =[]

  for i in range(quantidade_linhas):
    nova_linha = []

    for j in range(quantidade_colunas):
      nova_linha.append(0)


    custos_acumulados.append(nova_linha)

  custos_acumulados[0][0] = grade[0][0]

  for j in range(1, quantidade_colunas):
    custos_acumulados[0][j] = ( custos_acumulados[0][j - 1]+grade[0][j])

  for i in range(1, quantidade_linhas):
    custos_acumulados[i][0] = ( custos_acumulados[i-1][0]+grade[i][0])

  for i in range(1, quantidade_linhas):

    for j in range(1, quantidade_colunas):
       custos_cima = custos_acumulados[i-1][j]
       custos_esquerda = custos_acumulados[i][j-1]


       if custos_cima < custos_esquerda:
         menor_caminho = custos_cima
       else:
         menor_caminho = custos_esquerda

       custos_acumulados[i][j] = (menor_caminho + grade [i][j])




  return custos_acumulados[quantidade_linhas - 1][quantidade_colunas-1]


# Testes
grade1 = [
    [1, 3, 1],
    [1, 5, 1],
    [4, 2, 1]
]
print(menor_custo(grade1))  # 7

grade2 = [
    [1, 2],
    [1, 1]
]
print(menor_custo(grade2))  # 3
````

## Q6 — SQL | Censo Escolar

Execute a célula de setup abaixo para criar e popular o banco de dados. Depois responda cada subitem com uma query SQL.

### Modelo de Dados

```
ESCOLA                    TURMA                      MATRICULA
─────────────────         ──────────────────         ──────────────────────
id_escola    (PK)    ←──  id_escola                  id_matricula  (PK)
nome_escola               id_turma     (PK)     ←──  id_turma
municipio                 serie                       id_aluno
uf                        etapa_ensino                situacao_aprovado
rede                      qt_alunos_mat               (1 = aprovado,
localizacao                                            0 = reprovado/abandono)
(Urbana/Rural)       PROFESSOR_TURMA
                          ──────────────────     PROFESSOR
                          id_turma          ←──  ──────────────────
                          id_professor      ──→  id_professor  (PK)
                                                 nome_professor
                                                 formacao
                                                 (graduacao / especializacao
                                                  / mestrado / doutorado)
```

````python
# Setup do banco de dados — execute sem modificar
import sqlite3
import random

random.seed(42)
conn = sqlite3.connect(":memory:")
cur  = conn.cursor()

cur.executescript("""
CREATE TABLE escola (
    id_escola    INTEGER PRIMARY KEY,
    nome_escola  TEXT,
    municipio    TEXT,
    uf           TEXT,
    rede         TEXT,
    localizacao  TEXT
);
CREATE TABLE turma (
    id_turma      INTEGER PRIMARY KEY,
    id_escola     INTEGER REFERENCES escola(id_escola),
    serie         INTEGER,
    etapa_ensino  TEXT,
    qt_alunos_mat INTEGER
);
CREATE TABLE professor (
    id_professor   INTEGER PRIMARY KEY,
    nome_professor TEXT,
    formacao       TEXT
);
CREATE TABLE professor_turma (
    id_turma     INTEGER REFERENCES turma(id_turma),
    id_professor INTEGER REFERENCES professor(id_professor),
    PRIMARY KEY (id_turma, id_professor)
);
CREATE TABLE matricula (
    id_matricula      INTEGER PRIMARY KEY,
    id_turma          INTEGER REFERENCES turma(id_turma),
    id_aluno          INTEGER,
    situacao_aprovado INTEGER
);
""")

ufs    = ['DF','GO','MG','SP','BA','AM','CE','RS','PA','MT']
redes  = ['Federal','Estadual','Municipal','Privada']
locs   = ['Urbana','Rural']
etapas = ['Fundamental I','Fundamental II','Ensino Médio']
forms  = ['graduacao','especializacao','mestrado','doutorado']

for i in range(50):
    uf = random.choice(ufs)
    cur.execute("INSERT INTO escola VALUES (?,?,?,?,?,?)",
        (i+1, f'Escola {i+1:02d}', f'Município {uf}{i%10}', uf,
         random.choice(redes), random.choice(locs)))

for i in range(80):
    cur.execute("INSERT INTO professor VALUES (?,?,?)",
        (i+1, f'Professor {i+1:03d}', random.choice(forms)))

tid = 1
for eid in range(1, 51):
    for _ in range(random.randint(2, 6)):
        etapa = random.choice(etapas)
        serie = random.randint(1, 5 if 'Fundamental' in etapa else 3)
        qt    = random.randint(20, 45)
        cur.execute("INSERT INTO turma VALUES (?,?,?,?,?)",
            (tid, eid, serie, etapa, qt))
        for _ in range(random.randint(1, 3)):
            pid = random.randint(1, 80)
            cur.execute("INSERT OR IGNORE INTO professor_turma VALUES (?,?)", (tid, pid))
        for aluno in range(1, qt + 1):
            aprovado = 1 if random.random() > 0.18 else 0
            cur.execute("INSERT INTO matricula VALUES (?,?,?,?)",
                (tid * 100 + aluno, tid, aluno, aprovado))
        tid += 1

conn.commit()
print("Banco criado com sucesso!")
print(f"Escolas   : {cur.execute('SELECT COUNT(*) FROM escola').fetchone()[0]}")
print(f"Turmas    : {cur.execute('SELECT COUNT(*) FROM turma').fetchone()[0]}")
print(f"Matrículas: {cur.execute('SELECT COUNT(*) FROM matricula').fetchone()[0]}")
````

### Q6c
Liste as 5 escolas com maior média de alunos por turma. Mostre nome, município, UF e a média arredondada com uma casa decimal.

````python
# SUA RESPOSTA AQUI — Q6c
query_6c = """

SELECT
    e.nome_escola,
    e.municipio,
    e.uf,
    ROUND(SUM(t.qt_alunos_mat) * 1.0 / COUNT(t.id_turma),1)
      AS media_alunos_por_turma
FROM escola e
JOIN turma t ON e.id_escola = t.id_escola
GROUP BY e.id_escola, e.nome_escola, e.municipio, e.uf
ORDER BY media_alunos_por_turma DESC

LIMIT 5;

"""
pd.read_sql(query_6c, conn)
````

### Q6d
Há diferença na taxa de aprovação entre escolas Urbanas e Rurais? Monte uma query que mostre essa comparação por UF e localizacao.

````python
# SUA RESPOSTA AQUI — Q6d
query_6d = """

SELECT
    e.uf,
    e.localizacao,
    ROUND(AVG(m.situacao_aprovado) * 100, 1)
        AS taxa_aprovacao

FROM escola e
JOIN turma t ON e.id_escola = t.id_escola
JOIN matricula m ON t.id_turma = m.id_turma

GROUP BY e.uf, e.localizacao
ORDER BY e.uf, e.localizacao

"""
pd.read_sql(query_6d, conn)
````

Professores com doutorado ensinam em turmas com maior taxa de aprovação do que professores com apenas graduação? Monte a query comparando as quatro formações e interprete o resultado em no máximo três linhas.

````python
# SUA RESPOSTA AQUI — Q6e (query)
query_6e = """

SELECT
    p.formacao,
    ROUND(100.0 * SUM(m.situacao_aprovado) / COUNT(*),2)
      AS taxa_aprovacao
FROM professor p

JOIN professor_turma pt ON p.id_professor = pt.id_professor
JOIN turma t ON pt.id_turma = t.id_turma
JOIN matricula m ON t.id_turma = m.id_turma

GROUP BY p.formacao
ORDER BY taxa_aprovacao DESC;

"""
pd.read_sql(query_6e, conn)
````

Responda com suas próprias palavras:

2. O que é o desvio padrão? O que significa um desvio padrão alto no contexto do tempo de espera em uma UBS?

**Resposta Q11a** — escreva aqui:

1. A média é a soma de todos os valores dividida pela quantidade de dados, a mediana é o valor central quando os dados estão ordenados. A mediana é mais representativa que a média em distribuições assimétricas ou com outliers, poeque não é afetada por valores extremos.

2. O desvio padrão indica o quanto os valores de um conjunto de dados se afastam da média, nesse contexto de UBS o desvio padrao alto significa que ha grande variaçao nos tempos de atendimento

3. O outlier é um valor fora do padrão, muito abaixo ou muito acima dos bigodes do boxplot. É possível identificá-los visualmente como os pontos "bolinhas" que aparecem fora dos limites dos bigodes.

### Q11b — Prática

1. Calcule a média, a mediana e o desvio padrão do tempo de espera geral (desconsidere os valores ausentes).
2. Identifique e conte os outliers de tempo de espera usando o critério IQR: são outliers os valores abaixo de `Q1 − 1.5 × IQR` ou acima de `Q3 + 1.5 × IQR`.
3. Gere um histograma do tempo de espera e interprete a distribuição: ela é simétrica ou assimétrica? Qual é a direção da cauda?

````python
# SUA RESPOSTA AQUI — Q11b itens 1 e 2

# 1
tempo = df_sus['tempo_espera_min'].dropna()

media = tempo.mean()
mediana = tempo.median()
desvio_padrao = tempo.std()

print("media:", media)
print("mediana:", mediana)
print("desvio padrao:", desvio_padrao)


# 2.
Q1 = tempo.quantile(0.25)
Q3 = tempo.quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = tempo[(tempo < limite_inferior) | (tempo > limite_superior)]

print("Qtd outliers:", outliers.count())
````
