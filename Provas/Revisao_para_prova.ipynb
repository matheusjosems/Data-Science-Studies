# **Slide 4** --> **BeautfulSoup**

transforma HTML e XML em uma árvore de objetos Python , combinada com requests

- Encontrar elementos por **tag**, **classe**, **id** e outros atributos.
- Navegar na árvore do HTML (pais, filhos e irmãos).
- Extrair texto, links, tabelas e listas.
- Tratar HTML “malformado” de forma tolerante.

```python
# BeautifulSoup(html, "html.parser")
# Cria o objeto principal de parsing (soup), que permite navegar e buscar elementos no HTML

# find()
# Busca o PRIMEIRO elemento que corresponde aos critérios (tag, classe, id, etc)
# Ex: soup.find("a") -> retorna o primeiro <a>

# find_all()
# Busca TODOS os elementos que correspondem aos critérios
# Retorna uma lista de tags
# Ex: soup.find_all("p") -> retorna todos os <p>

# select()
# Busca elementos usando seletores CSS (igual no CSS/HTML)
# Retorna uma lista
# Ex: soup.select(".classe"), soup.select("#id"), soup.select("div > p")

# NavigableString
# Representa um TEXTO dentro de uma tag (não é uma tag, é só o conteúdo textual)
# Ex: <p>Olá</p> -> "Olá" é um NavigableString

# soup.find(...) e soup.find_all(...)
# Métodos principais de busca por tags no HTML

# soup.select(...)
# Método de busca usando seletores CSS (mais flexível e poderoso)

# .get_text(strip=True)
# Extrai o texto da tag removendo espaços extras no começo e fim
# Ex: <p>  Olá  </p> -> "Olá"

# tag["href"]
# Acessa diretamente um atributo da tag (pode dar erro se não existir)

# tag.get("href")
# Acessa o atributo com segurança (retorna None se não existir)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com" # Cria a variável url (tipo string).
html_exemplo = requests.get(url, timeout=10).text # Cria a variável html e armazena o código-fonte bruto do site.
soup_html = BeautifulSoup(html_exemplo, "html.parser") # Converte o HTML bruto em uma árvore de objetos Python e atribui à variável soup.

# pegar todos os links
for teste in soup_html.select("a[href]"): # A cada volta, o 'for' atribui à variável 'teste' um objeto Tag (referente ao <a>) que já foi criado dentro do soup.
    texto = teste.get_text(strip=True) # A variável 'teste' (que é um objeto) executa o método de extrair texto.
    variavel_href = teste.get("href") # A variável 'teste' acessa o seu atributo de link.
    print(texto, variavel_href)
```
